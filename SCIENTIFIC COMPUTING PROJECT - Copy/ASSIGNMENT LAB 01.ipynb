```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#displaying my dataset
c=pd.read_csv('all-ages.csv')
c
```

```python
#slicing data to reduce the number of entries to 100
CROPPED_DATA=c.head(100)
CROPPED_DATA
```

## SCATTER PLOT

```python
#SCATTER PLOT two display the relationship between the employed students and the number of students who take the course
x=CROPPED_DATA.Total
y=CROPPED_DATA.Employed
colors=(0,0,0)
area = np.pi*5
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('SCATTER PLOT SHOWING RELATIONSHIP BETWEEN EMPLOYED STUDENTS AND THE TOTAL STUDENTS WHO TAKE THE MAJOR')
plt.xlabel('TOTAL STUDENTS')
plt.ylabel('EMPLOYED STUDENTS')

plt.show()
```

## POLYNOMIAL FITS

```python
#plotting polynomial fits
x=CROPPED_DATA['Employed_full_time_year_round'].values
y=CROPPED_DATA['Total'].values
#plot the polynomials and points
p4=np.polyfit(x,y,4)
p4=np.polyfit(x,y,4)
#creating an abstraction for the math operation
xp4=np.poly1d(p4)
xp2=np.poly1d(p4)
xx=np.linspace(0,100,4)
#plot the polynomials and points
plt.plot(xx,np.polyval(p4,xx),"r*",label='degree 4')
plt.plot(x,y,'bo',markersize=2)
plt.legend(loc='upper left')
plt.title('Polynomial Fitting')
```

### NORMAL DISTRIBUTION

```python

mu=np.mean(x)
sigma=np.std(x)
x_array=np.linspace(0,100,100)
plt.plot(x_array,(1/np.sqrt(2*np.pi*sigma**2))*np.exp(-(x_array-mu)**2/(2*sigma**2)),'k--',linewidth=3)
plt.xlabel('')
plt.ylabel('')
plt.title('Histogram with $\mu={}, \quad\sigma={}$'.format(mu,sigma))
# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```
